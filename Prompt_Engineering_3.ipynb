```python
# ================================
# Exercise 3 – Self-Reflection Prompt (Meal Plan Example)
# ================================

import re
import difflib

# --------------------------------
# 1️⃣ BEFORE PROMPT
# --------------------------------
before_prompt = "Make a week long meal plan."

# --------------------------------
# 2️⃣ SELF-REFLECTION PROMPT (Improved Prompt)
# --------------------------------
self_reflection_prompt = """
Review your previous 7-day meal plan and evaluate it using the following criteria:

1) Fasting Requirement: Must support a 12-hour morning fast.
2) Protein Target: Each day must include at least 180g of protein.
3) Dinner Variety: Dinners must not repeat and must include diverse protein sources.
4) Sweet Treat: Each day must include a controlled sweet treat.
5) Practicality: Meals must be realistic and easy to prepare.
6) Formatting: Clearly structured by day with estimated protein totals.

Identify weaknesses and then provide a fully revised 7-day meal plan that satisfies all constraints.
"""

# --------------------------------
# 3️⃣ BEFORE OUTPUT (Paste your original meal plan here)
# --------------------------------
before_output = """
PASTE YOUR ORIGINAL MEAL PLAN HERE
"""

# --------------------------------
# 4️⃣ AFTER OUTPUT (Paste your improved meal plan here)
# --------------------------------
after_output = """
PASTE YOUR REVISED MEAL PLAN HERE
"""

# --------------------------------
# Helper Functions
# --------------------------------
def count_days(text):
    return len(re.findall(r'Day\s*\d', text))

def protein_mentions(text):
    return re.findall(r'\d+\s*g', text)

# --------------------------------
# PRINT STRUCTURED SUBMISSION
# --------------------------------
print("========== EXERCISE 3 SUBMISSION ==========\n")

print("1) BEFORE PROMPT:\n")
print(before_prompt)

print("\n2) BEFORE OUTPUT:\n")
print(before_output)

print("\n3) SELF-REFLECTION PROMPT USED:\n")
print(self_reflection_prompt)

print("\n4) AFTER (REVISED) OUTPUT:\n")
print(after_output)

print("\n5) STRUCTURE CHECKS:")
print(f"Days in AFTER plan detected: {count_days(after_output)}")
print(f"Protein values mentioned in AFTER plan: {protein_mentions(after_output)}")

print("\n6) CHANGES (BEFORE → AFTER):\n")
diff = difflib.unified_diff(
    before_output.splitlines(),
    after_output.splitlines(),
    fromfile="Before Plan",
    tofile="After Plan",
    lineterm=""
)

print("\n".join(diff))

print("\n========== END ==========")
```
